```python
#libraries needed for this analysis
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
```

```python
#import data from CSV file by creating pathway
# The path to our CSV file
imdb_df = Path("cleaned_imdb_file.csv")

# Read our Crowdfunding data into pandas
imdb_df = pd.read_csv(imdb_df)
imdb_df.head()
```

```python
'Filtering out Drama Movies'
```

```python
#Creating a df with only drama movies
drama_movies = imdb_df[imdb_df["Genre"].str.contains("Drama")]
drama_movies.head()
```

```python
'Drama Movie Runtimes and Meta-Score'
```

```python
#Scatterplot of Runtimes vs Meta-Score
drama_runtimes = drama_movies.plot.scatter(x='Runtime (in mins)',

                      y='Meta-Score')
```

```python
'Runtime vs Meta-Score'
```

```python
#Line of best fit
x = drama_movies['Runtime (in mins)']
y = drama_movies['Meta-Score']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
runtime_regression = x * slope + intercept
line_bf = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
plt.scatter(x,y)
plt.plot(x,runtime_regression,"r-")
plt.xlabel('Runtime (in mins)')
plt.ylabel('Meta-Score')
print(line_bf)
```

```python
'Distribution of Runtimes'
```

```python
#Histogram of Runtime
drama_movies['Runtime (in mins)'].hist()
plt.xlabel('Runtime (in mins)')
plt.ylabel('Frequency')
plt.title('Distribution of Runtime (in mins)')
```

```python
#Histogram of Meta-Score
drama_movies['Meta-Score'].hist()
plt.xlabel('Meta-Score')
plt.ylabel('Frequency')
plt.title('Distribution of Meta-Score')
```

```python
'Removal of Outliers to Create a Normal Distribution'
```

```python
# Parameters for outlier removal
mean = drama_movies['Runtime (in mins)'].mean()
std_dev = drama_movies['Runtime (in mins)'].std()

# Define a criteria for outlier removal
lower_bound = mean - 2 * std_dev
upper_bound = mean + 2 * std_dev

# Filter the DataFrame to remove outliers
normal_dist_df = drama_movies[(drama_movies['Runtime (in mins)'] >= lower_bound) & (drama_movies['Runtime (in mins)'] <= upper_bound)]

# Generate random numbers following a normal distribution with the same mean and standard deviation
normal_distribution = np.random.normal(mean, std_dev, normal_dist_df.shape[0])

# Create a new DataFrame with the normal distribution data
normal_dist_df = pd.DataFrame(normal_distribution, columns=['Runtime (in mins)'])

# Display the resulting DataFrame
normal_dist_df['Runtime (in mins)'].hist()
plt.xlabel('Runtime (in mins)')
plt.ylabel('Frequency')
plt.title('Distribution of Runtime (in mins)')
```

```python
# Generate Meta-Score values based on a normal distribution
mean_meta_score = drama_movies['Meta-Score'].mean()
std_dev_meta_score = drama_movies['Meta-Score'].std()
meta_score_distribution = np.random.normal(mean_meta_score, std_dev_meta_score, normal_dist_df.shape[0])

# Add the generated Meta-Score data to the DataFrame
normal_dist_df['Meta-Score'] = meta_score_distribution

# Display the resulting DataFrame
normal_dist_df['Meta-Score'].hist()
plt.xlabel('Meta-Score')
plt.ylabel('Frequency')
plt.title('Distribution of Meta-Score')
```
